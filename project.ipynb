## Functions to call Deezer API Endpoints

```python
import requests

def get_artist_data(artist_id):
    # Deezer API endpoint for artist details
    api_url = f"https://api.deezer.com/artist/{artist_id}"

    try:
        # Send GET request to the API endpoint
        response = requests.get(api_url)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the JSON data from the response
            artist_data = response.json()
            return artist_data
        else:
            # Print an error message if the request was not successful
            print(f"Error: Unable to fetch data. Status Code: {response.status_code}")
            return None

    except Exception as e:
        print(f"Error: {e}")
        return None
    
import requests

def get_artist_albums(artist_id):
    # Deezer API endpoint for artist details
    api_url = f"https://api.deezer.com/artist/{artist_id}/albums"

    try:
        # Send GET request to the API endpoint
        response = requests.get(api_url)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the JSON data from the response
            artist_albums_data = response.json()
            return artist_albums_data
        else:
            # Print an error message if the request was not successful
            print(f"Error: Unable to fetch data. Status Code: {response.status_code}")
            return None

    except Exception as e:
        print(f"Error: {e}")
        return None

import requests

def get_album_tracks(album_id):
    # Deezer API endpoint for artist details
    api_url = f"https://api.deezer.com/album/{album_id}/tracks"

    try:
        # Send GET request to the API endpoint
        response = requests.get(api_url)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the JSON data from the response
            tracks_data = response.json()
            return tracks_data
        else:
            # Print an error message if the request was not successful
            print(f"Error: Unable to fetch data. Status Code: {response.status_code}")
            return None

    except Exception as e:
        print(f"Error: {e}")
        return None

import requests

def get_album_genres(album_id):
    # Deezer API endpoint for artist details
    api_url = f"https://api.deezer.com/album/{album_id}/tracks"

    try:
        # Send GET request to the API endpoint
        response = requests.get(api_url)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the JSON data from the response
            tracks_data = response.json()
            return tracks_data
        else:
            # Print an error message if the request was not successful
            print(f"Error: Unable to fetch data. Status Code: {response.status_code}")
            return None

    except Exception as e:
        print(f"Error: {e}")
        return None
```

# Funtions to call MusixMatch API methods

```python
import requests

def get_lyrics_helper(api_key, track_id):
    base_url = "https://api.musixmatch.com/ws/1.1/"
    endpoint = "track.lyrics.get"

    params = {
        'apikey': api_key,
        'track_id': track_id,
        'format': 'json',
        
    }

    response = requests.get(f"{base_url}{endpoint}", params=params)
    data = response.json()
    return data['message']['body']['lyrics']['lyrics_body']

    # if 'lyrics' in data['message']['body']:
    #     lyrics_body = data['message']['body']['lyrics']['lyrics_body']
    #     # Check if the lyrics are not instrumental
    #     if "Instrumental" not in lyrics_body:
    #         return lyrics_body
    #     else:
    #         return "Lyrics found but marked as instrumental."
    # else:
    #     return "Lyrics not found."

def get_lyrics(api_key, track_title, artist_name):
    base_url = "https://api.musixmatch.com/ws/1.1/"
    endpoint = "matcher.track.get"
    
    params = {
        'apikey': api_key,
        'q_track': track_title,
        'q_artist': artist_name,
        'format': 'json',
    }

    response = requests.get(f"{base_url}{endpoint}", params=params)
    data = response.json()
    id = data['message']['body']['track']['track_id']
    lyrics = get_lyrics_helper(api_key, id)
    return lyrics.split("...")[0].strip()

def get_track_genre(api_key, track_title, artist_name):
    base_url = "https://api.musixmatch.com/ws/1.1/"
    endpoint = "matcher.track.get"
    
    params = {
        'apikey': api_key,
        'q_track': track_title,
        'q_artist': artist_name,
        'format': 'json',
    }

    response = requests.get(f"{base_url}{endpoint}", params=params)
    data = response.json()
    return data['message']['body']['track']['primary_genres']['music_genre_list'][0]['music_genre']['music_genre_name'].split('/')[0]

# Replace with Musixmatch API key
musixmatch_api_key = "b9004adfeacd194b70efce1f07424f75"

# Example usage
title = "coming back for you"
artist = "maroon 5"
lyrics = get_lyrics(musixmatch_api_key, title, artist)
#lyrics = get_lyrics(musixmatch_api_key, id)
print(lyrics)
```

Function to append CSV

```python
import csv

def append_to_csv(data_dict):

    # Sample dictionary to append to the CSV file
    data_to_append = data_dict

    # CSV file path
    csv_file_path = "SongsData.csv"

    # Open the CSV file in append mode
    with open(csv_file_path, 'a', newline='', encoding='utf-8') as csvfile:
        # Create a CSV writer object
        csv_writer = csv.DictWriter(csvfile, fieldnames=data_to_append.keys())

        # Check if the file is empty (write header if necessary)
        file_is_empty = csvfile.tell() == 0
        if file_is_empty:
            csv_writer.writeheader()

        # Write the data to the CSV file
        csv_writer.writerow(data_to_append)
```
