```python
import nltk
import random
import pandas as pd
nltk.download('punkt')
```

```python
# for hip hop
df =  pd.read_csv('SongsData.csv')
lyrics_data = ''
for index, lyrics in df.iterrows():
    if lyrics['Genre'] == 'Hip Hop':
        lyrics_data += str(lyrics['Lyrics'])
lyrics_data
lyrics_data.replace('(','')
```

```python
# Tokenize the lyrics
tokens = nltk.word_tokenize(lyrics_data)
print(tokens)
```

```python
# Choose the N-gram order (e.g., N=2 for bigrams)
N = 2
ngrams = list(nltk.ngrams(tokens, N))
ngrams[0]
```

```python
# Build a frequency table
freq_table = nltk.FreqDist(ngrams)
freq_table.most_common()
```

```python
# Generate new lyrics
seed = ""
count = 0
generated_lyrics = seed

while True:
    prev_words = nltk.word_tokenize(generated_lyrics)[-N+1:]
    next_word_candidates = [ngram[-1] for ngram in freq_table if ngram[:-1] == tuple(prev_words)]

    if next_word_candidates and count < 200:
        # Calculate probabilities based on frequencies
        probabilities = [freq_table[ngram] / sum(freq_table.values()) for ngram in freq_table if ngram[:-1] == tuple(prev_words)]
        
        # Use weighted choice based on probabilities
        next_word = random.choices(next_word_candidates, weights = probabilities)[0]
        generated_lyrics += " " + next_word

        count += 1
    else:
        break

print(generated_lyrics)
```
